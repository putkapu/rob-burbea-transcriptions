```python
import os
import pandas as pd
import re
import urllib.request
from PyPDF2 import PdfReader
```

```python
def get_text_between_parentheses(string):
    text = re.findall(r'\(([^)]+)\)', string)
    if len(text) > 0:
        text = text[0]
    else:
        text = ""
        
    return text
```

```python
df = pd.read_csv("data/Spreadsheet-Grid view.csv")
df.head()
```

```python
df["Download Link"] = df['Transcript or Writing'].dropna().apply(get_text_between_parentheses)
```

```python
path = "data/original transcriptions/"
```

```python
# download files

for url, title in df[["Download Link", "Title of Talk or Writing"]].dropna(subset=["Download Link"]).values:
    # Download the file
    file = f"{title}.pdf".replace("/", "|")
    filepath = f{path}/{file}"
    urllib.request.urlretrieve(url, filepath)
```

```python
def get_text_around_word(word, main_text, n=50):
    result = ""
    if word in text:
        word_idx = main_text.index(word)
        
        start_idx = max(0, word_idx-n)
        result = main_text[start_idx:word_idx+n]
    
    return result
```

```python
files = os.listdir(path)
for f in files:
    fp = f"{path}/{f}"
    try:
        with open(fp, 'rb') as file:
            # create a PDF object
            reader = PdfReader(file)
            text = ""
            for page in reader.pages:
                text += page.extract_text() + "\n"

        text = text.lower()
        for word in ["serpent", "snake"]:
            if word in text:
                print(f"from: {f}")
                print(f">> {word}")
                print(f"{get_text_around_word(word, text, n=300)}")
                print()
            
    except:
        continue
        
```
